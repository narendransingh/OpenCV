```python
# overlay two images Image 3 and Image 4
```

```python
image3 = Image.open("sunset1.jpg")
image4 = Image.open("blow_up.jpg")
```

```python
image3.size
```

```python
image4.size
```

```python
image4_copy = image4.copy() #create a copy of image4 (bigger image)
image4_copy.paste(image3, (50,50)) # now we pasting image3 on image4_copy starting from 50, 50 pixel
image4_copy.save("overlay.jpg")
```

### Image Rotation

```python
image3.rotate(45) # rotates the image by 90 degree, rotate 45 degree will lose information
```

```python
image3.rotate(45, expand = True) #this to keep all the values of image.
```

```python
# Transpose flip image
```

```python
image3.transpose(Image.FLIP_LEFT_RIGHT) # mirror image
```

```python
image3.transpose(Image.FLIP_TOP_BOTTOM)
```

```python
# image to greyscale
grey_image = image3.convert("L") # converts to greyleevel image can use RGB to convert to color
grey_image
```

```python
#convert series of images in folder

import glob
```

```python
path = "/resources/labs/*.jpg" # all the images in jppeg
```

```python
for file in glob.glob(path):
    print(file)
    a = Image.open(file) # open list of file, do check it opens one at a time
    rotated45 = a.rotate(45, expand = True) #rotate file for example
    rotated45.save(file+"_rotated.png", "PNG") # save each file 
```
