# Object Detection in OpenCV

Indroduction to look into Hue, Saturation and Value(HSV), these are luminus value and plays important information whne detecting objects in images.

HDV as heckthon model (cone model). Hue is bottom of cone ranges from 0 - 360. Diff color can be selected based on ange. these color are red yellow cyan blue megenta.

Saturation is depth of pigment or dominace of hue, starts from center to moves toward outer layer of the cone. Range of Saturation is 0 - 1. (0 - 100%)

Value is brightness is of the color as move from tip of the cone to bottom of come. value ranges - 0 - 1 (0 - 100%)

```python
#detect image with colored smarties

def nothing(x):
    pass


##naming the image window, since is difficult to detect the lower and upper bound of color we will use tracker
cv.namedWindow("Tracking") 


while True:
    
    frame = cv.imread('smarties3.jpg')
    
    hsv = cv.cvtColor(frame, cv.COLOR_BGR2HSV) # property we using to convert to HSV
    
    #Lets restrict for now to limit threshold of HSV image beteween lower blue color and upper blue colors.
    #We can find the actual HSV value using the Tracking.
    
    lower_blue = np.array([110, 50, 50]) #HSV value for lower limit blue color
    upper_blue = np.array([130, 255, 255]) #HSV value for upper limit blue color 
    
    mask = cv.inRange(hsv, lower_blue, upper_blue) #we will threshold the HSV image to get only blue color
    
    res = cv.bitwise_and(frame, frame, mask=mask) #and opertation value:(src1, src2, mask(mask variable we created))
    
    
    cv.imshow('frame', frame)
    cv.imshow('mask', mask)
    cv.imshow('result', res)
    
    key = cv.waitKey(1)
    if key == 27: #27 key is escape key
        break

        
cv.destroyAllWindows()
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)   
```

```python
#Will use tracker to identify the hue, saturation and value for upper and lower bound. Will have four windows,
#including the Tracking which will give estimate value of 6 parameters.


def nothing(x):
    pass


##naming the image window, since is difficult to detect the lower and upper bound of color we will use tracker
cv.namedWindow("Tracking") 

#create tracker values:(trackerName, nameofwindow, startvalue, endvalue and callback function)
cv.createTrackbar('LH', 'Tracking', 0, 255, nothing)  # for Lower Hue
cv.createTrackbar('LS', 'Tracking', 0, 255, nothing)  #for lower Saturation
cv.createTrackbar('LV', 'Tracking', 0, 255, nothing)  # from Lower value
cv.createTrackbar('UH', 'Tracking', 255, 255, nothing)  #for uper Hue
cv.createTrackbar('US', 'Tracking', 255, 255, nothing)  #for upper Saturation
cv.createTrackbar('UV', 'Tracking', 255, 255, nothing)  #for upper value



while True:
    
    frame = cv.imread('smarties3.jpg')
    
    hsv = cv.cvtColor(frame, cv.COLOR_BGR2HSV) # property we using to convert to HSV
    
    #get trackbar in image name Tracking.
    
    l_h = cv.getTrackbarPos('LH', 'Tracking')
    l_s = cv.getTrackbarPos('LS', 'Tracking')
    l_v = cv.getTrackbarPos('LV', 'Tracking')
    u_h = cv.getTrackbarPos('UH', 'Tracking')
    u_s = cv.getTrackbarPos('US', 'Tracking')
    u_v = cv.getTrackbarPos('UV', 'Tracking')
    
    
    
    
    #Lets restrict for now to limit threshold of HSV image beteween lower blue color and upper blue colors.
    #NOw we provide the value of from get tracker to threshold
    
    lower_blue = np.array([l_h, l_s, l_v]) #HSV value for lower limit blue color
    upper_blue = np.array([u_h, u_s, u_v]) #HSV value for upper limit blue color 
    
    mask = cv.inRange(hsv, lower_blue, upper_blue) #we will threshold the HSV image to get only blue color
    
    res = cv.bitwise_and(frame, frame, mask=mask) #and opertation value:(src1, src2, mask(mask variable we created))
    
    
    cv.imshow('frame', frame)
    cv.imshow('mask', mask)
    cv.imshow('result', res)
    
    key = cv.waitKey(1)
    if key == 27: #27 key is escape key
        break

        
cv.destroyAllWindows()
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)   
```

```python
#for Video detection:
#Use the tracker to select the hue, saturation or the value to detect any color.

def nothing(x):
    pass



cap = cv.VideoCapture(0); #capture video from webcamera of laptop.



##naming the image window, since is difficult to detect the lower and upper bound of color we will use tracker
cv.namedWindow("Tracking") 

#create tracker values:(trackerName, nameofwindow, startvalue, endvalue and callback function)
cv.createTrackbar('LH', 'Tracking', 0, 255, nothing)  # for Lower Hue
cv.createTrackbar('LS', 'Tracking', 0, 255, nothing)  #for lower Saturation
cv.createTrackbar('LV', 'Tracking', 0, 255, nothing)  # from Lower value
cv.createTrackbar('UH', 'Tracking', 255, 255, nothing)  #for uper Hue
cv.createTrackbar('US', 'Tracking', 255, 255, nothing)  #for upper Saturation
cv.createTrackbar('UV', 'Tracking', 255, 255, nothing)  #for upper value



while True:
    
    _, frame = cap.read()
    
    hsv = cv.cvtColor(frame, cv.COLOR_BGR2HSV) # property we using to convert to HSV
    
    #get trackbar in image name Tracking.
    
    l_h = cv.getTrackbarPos('LH', 'Tracking')
    l_s = cv.getTrackbarPos('LS', 'Tracking')
    l_v = cv.getTrackbarPos('LV', 'Tracking')
    u_h = cv.getTrackbarPos('UH', 'Tracking')
    u_s = cv.getTrackbarPos('US', 'Tracking')
    u_v = cv.getTrackbarPos('UV', 'Tracking')
    
    
    
    
    #Lets restrict for now to limit threshold of HSV image beteween lower blue color and upper blue colors.
    #NOw we provide the value of from get tracker to threshold
    
    lower_blue = np.array([l_h, l_s, l_v]) #HSV value for lower limit blue color
    upper_blue = np.array([u_h, u_s, u_v]) #HSV value for upper limit blue color 
    
    mask = cv.inRange(hsv, lower_blue, upper_blue) #we will threshold the HSV image to get only blue color
    
    res = cv.bitwise_and(frame, frame, mask=mask) #and opertation value:(src1, src2, mask(mask variable we created))
    
    
    cv.imshow('frame', frame) # open original frame
    cv.imshow('mask', mask)   # mask framw which display the changes using hue, saturation, and value
    cv.imshow('result', res)  # result using AND gate which provides the common region between two frames with mask
    
    key = cv.waitKey(1)
    if key == 27: #27 key is escape key
        break


cap.release()      # release all the video frames  

cv.destroyAllWindows()
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1) 

```
