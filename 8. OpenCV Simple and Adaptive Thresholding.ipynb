## Simple Image Thresholding and Adaptive Thresholding

Thresholding is popular segmentation technique to seperating an object from its backgroud. Process of thresholding requries comparing each pixel of an image with predefined threshold value. 

This divides the pixels of input image into two group: 1. involve pixels having intensity valuve of pixels lower than threshold values 2.involve pixels having intensity of pixels greater than threshold values

Using different thresholding techiques in OpenCV, we can give different values to these pixels which has higher than the threshold values or lower threshold values.


```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
image = cv.imread('linear_gradient.png', 1)
```

```python
plt.imshow(image)
```

```python
cv.imshow('image_color', image)

cv.waitKey(0)
cv.destroyAllWindows()
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
```

```python
#result of threshold returen 2 values, ret value and threshold value of image

#value(image, threshold min value(), max value of thres, thres type)  

_,threshold1 = cv.threshold(image,127, 255, cv.THRESH_BINARY)  
```

```python
cv.imshow('thres_image_binary', threshold1)

cv.waitKey(0)
cv.destroyAllWindows()
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
```

#### Binary Thres
we can see comparing every pixel in original image to 127 and if value if pixel less than 127 value is assigned to 0(black), and if value greater than 127 the pixel value is 255 i.e(white) 

```python
plt.imshow(threshold1) 
```

```python
#Thres Binary Inverse

#This will is inverse of thres_binary, if pixel value less than 127 then thres value is 1(white), else 0(black)

_,threshold2 = cv.threshold(image,127, 255, cv.THRESH_BINARY_INV) 
```

```python
plt.imshow(threshold2)
```

```python
#Thres Trunk
#upto the threshold the value wont be change, i.e till 0 to 127 pixel. after pixel 127 then value will remain 127.

_,threshold3 = cv.threshold(image,127, 255, cv.THRESH_TRUNC) 
```

```python
plt.imshow(threshold3)
```

```python
#Thres Twozero
#Whenever pixel value is less than thres_value i.e. 127 the value is zero(black), 
#and greater than thres_value 127 the image will remain the 'same'

_,threshold4 = cv.threshold(image,127, 255, cv.THRESH_TOZERO) 
```

```python
plt.imshow(threshold4)
```

```python
#Tres_toZero_Inverse
#Inverse of above threshold, initial value before the threshold_value(127), image to remain same, but greater than 
#thres_value (127) will change the image zero(black)

_,threshold5 = cv.threshold(image,127, 255, cv.THRESH_TOZERO_INV) 
```

```python
plt.imshow(threshold5)
```

### Using Adaptive Threshold

Adaptive Threshold is method where threshold value is calculated for smaller region, here threshold value is not global like in Simple Threshold. There will be different thresholding values for different region.

Why we need, using simple may not be good idea, like image has different lighting condition at differnet areas of image. 

It calculates threshold values for smaller region if images, diff threshold value for same image as a result, it gives better solution with varying luminiation.

```python
image2 = cv.imread('Suduko2.jpg', 0) #have to read image in greyscale for AdaptiveThreshold
```

```python
plt.imshow(image2)
```

```python
# using simple threshold (binary threshold)

_,thres_binary = cv.threshold(image2, 127, 255, cv.THRESH_BINARY) #threhold value is global. 
```

```python
plt.imshow(thres_binary) # image has different illumination at different place
```

### Using Adaptive Threshold

##### adaptive thres values:(source, maxvalue(non-zero value assigned to pixels for which condition is satisfied), adaptive method(how threshold value is calculated), thresholdtype method, block size for neighborhood area, C -constant which is subtracted from mean or weighted mean )

There are two methods for Adaptive
1. Threshold_Mean_C - thres value mean of the neighbor area.(block size
2. Thres_gaussian_c - thresh value is weighted sum of gaussian 

```python
#Using Adaptive Mean

thres_adpt = cv.adaptiveThreshold(image2, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 2);
```

```python
plt.imshow(thres_adpt)
```

```python
cv.imshow('thres_adaptive', thres_adpt)

cv.waitKey(0)
cv.destroyAllWindows()
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
```

```python
#using Gaussian Weighted Mean

thres_gau = cv.adaptiveThreshold(image2, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)
```

```python
cv.imshow('thres_gaussian', thres_gau)

cv.waitKey(0)
cv.destroyAllWindows()
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
```

```python
plt.imshow(thres_gau)
```
