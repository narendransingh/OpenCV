# Image Gradient

IG is a directional change in intesity or color if image. We can use to find edges in image. Some of different methods. 
1. Laplacian (Laplacian derivatives)
2. Sobel X(Gaussian and derivative)  
3. Sobel Y 

These are different gradient functions uses different mathematical calculation to create the gradient.

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv.imread('Suduko.png', 0)
```

```python
plt.imshow(image)
```

```python
#Laplacian Image Gradient: (find edges based on intesity change)

#values:(image, datatype(using 64bit float because negative slope induce by transforming white to black image), 
#        kernel)
#we then take the absolute value of laplacian and then convert it back to unsigned 8 bit integer to display

lap = cv.Laplacian(image, cv.CV_64F, ksize=3)
lap = np.uint8(np.abs(lap))
```

```python
#Sobel X and Sobel Y: (Global)
#values:(image, datatype, dx(order of derivative x - 1/0, dy(order of derivative y - 1/0), kernel size (optional)))
#convert it back to unsigned 8 bit integer to display
#X stands for derivative on x asix so horizontal
#Y stnads for derivatiove on y axis so vertical axis
```

```python
sobelx = cv.Sobel(image,cv.CV_64F, 1, 0 )
sobelx = np.uint8(np.abs(sobelx))
```

```python
sobely = cv.Sobel(image,cv.CV_64F, 0, 1)
sobely = np.uint8(np.abs(sobely))
```

```python
sobel_comb = cv.bitwise_or(sobelx,sobely)
```

```python
titles = ['Originial', 'Laplacian', 'SobelX', 'SobelY', 'Sobel_Combined']
images = [image, lap, sobelx, sobely, sobel_comb]
```

```python
for i in range(5):
    plt.subplot(2, 3, i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
    
```

# Canny Edge Detection

It uses multi-stage algorithm to detect wide range of edges in images. by John F Canny. it can be broken in 5 step

1. Gaussian to smooth, noise removal.
2. Find intensity gradient of image
3. Non-maximum supression, to get ride supress response
4. Double Threshold, to fetermone potential edges
5. Edge Tracking by Hysteresis, final only show strong edges and leave which are weak or not connected to strong edges 

We have builtin function in OPenCV

Its application is better than Image Gradient(sobels) or other edge dection, it has less noise.


```python
image2 = cv.imread('Lena_large.jpg', 0)
```

```python
plt.imshow(image2)
```

```python
#decalte variable canny
#values:(imagesrc, threshold1,threshold2 (we provide threshold for hysteresis process), )
#would wnat to add trackbar to changes in image.

canny = cv.Canny(image2, 100, 200)

```
