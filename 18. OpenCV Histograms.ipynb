## Histogram


it is graph give you intesity distribution of image, how they useful. Can find contrast, brigtness

Uses - can tell if image is properly exposed. lighting condition were harsh or flat. 

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = np.zeros((200,200), np.uint8)
cv.rectangle(image, (0,100), (200,200), (255), -1)
```

```python
plt.imshow(image)
```

### Different ways to find histogram of image

```python
#Histogram values:(imagesrc, maxpixelvalue, rangeofpixel)

plt.hist(image.ravel(), 256, [0, 256])
plt.show()

#the graph show 200x200 pixel(y axis), intensity(x axis) is all black(zeros) so all value on zero plot

# adding the rectangle to the image, the image is has changed to overall 20000 pixel black(@0) and 20000 white(@255)
```

```python
image1 = np.zeros((200,200), np.uint8)
cv.rectangle(image1, (0,100), (200,200), (255), -1)
cv.rectangle(image1, (0,50), (100,100), (127), -1)
```

```python
plt.imshow(image1)
```

```python
#Histogram values:(imagesrc, maxpixelvalue, rangeofpixel)

plt.hist(image1.ravel(), 256, [0, 256])
plt.show()


#You can see teh intensity at 127 is 100x50 pixels and similarly intensity 0(block) has total pixel 15000 and so on
```

```python
#Load actual image

image3 = cv.imread('Lena.png', 0) #greyscale
```

```python
plt.imshow(image3)
```

```python
#histogram

plt.hist(image3.ravel(), 256, [0, 256]) #for greyscale
plt.title('Histogram Lena grayscale')
plt.show()

#can observe the most amount of number pixel is at intensity 51
```

```python
#Load color image

image4 = cv.imread('Lena.png', 1) #color
```

```python
#split color image to bgr channel

b, g, r = cv.split(image4)
```

```python
#to view in each scale

cv.imshow('orig',image4)
cv.imshow('blue',b)
cv.imshow('green',g)
cv.imshow('red',r)


cv.waitKey(0)
cv.destroyAllWindows()
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
```

```python
plt.imshow(g)
```

```python
#Histogram for color image

plt.hist(b.ravel(), 256, [0, 256]) 
plt.hist(g.ravel(), 256, [0, 256]) 
plt.hist(r.ravel(), 256, [0, 256]) 
plt.title('Histogram Color Lena Image')
plt.show()

```

```python
#Histogram using OpenCV2

image5 = cv.imread('Lena.png', 0) #color
```

```python
#histogram
#values:(imgsrc(image to be in square bracket), index of channel(grayscale - 1 channel), 
#      imagemask(can give None since greyscale), histsize(representation of bin count), range(0 to 256))

hist = cv.calcHist([image5], [0], None, [256], [0, 256])
```

```python
plt.title('Histogram using OpenCV')
plt.plot(hist)
```
