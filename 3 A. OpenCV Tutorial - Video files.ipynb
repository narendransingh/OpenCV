```python
import cv2 as cv
```

```python
#cap = cv2.VideoCapture('filename.avi') # use videofile name
cap = cv.VideoCapture(0); # this to capture image from camera

while (True):
    # using cap instance we will call read, if return true. if frame available will save in frame variable
    ret, frame = cap.read() 
    
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY) #we change the default capture image BGR to grey
    
    cv.imshow('frame', gray)
    
    if cv.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
```

```python
#saving the captured image using VideoWriter. t

cap = cv.VideoCapture(0); #capture image

fourcc = cv.VideoWriter_fourcc(*'XVID') # can also provide like cv.VideoWriter_fourcc('X','V','I','D') use this.

#take argument name, 4cc code:4 byte code to specify video kodak,(www.fourcc.org/codecs.php), frames/sec, size of video
out = cv.VideoWriter('output.avi', fourcc, 20.0, (640,480)) 

while (True):
    # using cap instance we will call read, if return true. if frame available will save in frame variable
    ret, frame = cap.read() #ret is boolean
    
    if ret == True: #check if ret is true, only then we will ssave the file
    
        print(cap.get(cv.CAP_PROP_FRAME_HEIGHT))# to get different properties of the video, here the height of video
        print(cap.get(cv.CAP_PROP_FRAME_WIDTH)) # to get the width of the image, cand fid list from Opencv website
        
        #the output frame size from video camera is 720x1280 pixels
        
        
        out.write(frame) #save using method out is instance of video write
        
        gray = cv.cvtColor(frame, cv2.COLOR_BGR2GRAY) #we change the default capture image BGR to grey
    
        cv.imshow('frame', gray)
    
        if cv.waitKey(1) & 0xFF == ord('q'): # pressing q key on keyboard to quit.
            break
    else:
        break

cap.release()
out.release()
cv.destroyAllWindows()
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
```
