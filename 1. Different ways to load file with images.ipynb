```python
# using Skimage and try image processing
from skimage import io
from scipy import ndimage

# reference for scipy read online from http:/docs.scipy.org..../ndimage.html
```

```python
image = io.imread("./Negative/18476.jpg", as_grey=True) # import in greyscale image, easier to import in skimage

rotated = ndimage.rotate(image, 45) #by default the image is expand unlike in Pillow
rotated = ndimage.rotate(image, 45, reshape=False) # to cut the image if outbound retain original size
```

```python
#use different filters
gaussian_filtered_image = ndimage.gaussian_filter(image, Sigma=11) # will blur the image remove noise.
plt.imshow(gaussian_filtered_image) #gaussian does not preserve the edges

meadian_filtered_image = ndimage.median_filter(image, 3)
plt.imshow(meadian_filtered_image) # pereserve the edges for noisy image
```

```python
# to open mulitple files using matplotlib

flippedLR = np.fliplr(image) # using numpy to flip the image to mirror image
flippedUD =np.flipud(image) # using numpy filp image up to down

plt.subplot(2,2,1) #this to define number or rows, no of columns and number of images
plt.imshow(image, cmap='Greys') #to change image color style
plt.subplot(2,2,3)
plt.imshow(flippedLR, cmap='Blues') # will change image color style
plt.subplot(2,2,4)
plt.imshow(flippedUD) #this should display images all together
```
