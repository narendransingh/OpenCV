## Detect Corners using Shi Thomasi Corner Detection using OpenCV

It was created by Ja She and Shi Thomasi in 1994, making small change to Harris Corner in paper called Good Feature to Track.

it is similar to Harris Corner, but change in score of R is calculated. It gives better result than Harris.

Using Shi Thomasi Corner we can find top end corner, we can provide number of corners we want instead of detecting all the corners in the image.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread('shapes.jpg')
```

```python
plt.imshow(image)
```

```python
image_gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(image_gray)
```

#### Applying Shi Thomasi
values:(image(grayscale), max number of corners(only strongest corner will be returned), quality_level(minimal expected quality of image corner), min_distance(min Euclidean distance)

```python
#apply detection

corners = cv.goodFeaturesToTrack(image_gray, 22, 0.01, 10)
```

```python
#convert corners to integer value, int0 is alias to int64.

corners = np.int0(corners)
```

```python
#we iterate over all the corners to find X and Y values and draw circle on the image using X and Y

for i in corners:
    
    #get x and y values.
    x, y = i.ravel()
    
    #draw the circle on original image
    #values:(image, center, radius, color, thickness (-1 = fill circle))
    cv.circle(image, (x,y), 10, (0,255,0), -1)
```

```python
#display the image

cv.imshow('corners', image)
cv.waitKey(0)
cv.destroyAllWindows()
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
```
