# Basic operation on Images

Split on image, Resize, Merge, Region of Interest and Add two images with weights

```python
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
image = cv.imread('shankar_resized.jpeg', 1)
```

```python
image.shape #number of rows, number of columns, size of image channel.
```

```python
image.size #total number of pixels
```

```python
image.dtype #type of image
```

```python
plt.imshow(image)
```

```python
cv.imshow('shankar', image)

cv.waitKey(0)
cv.destroyAllWindows()
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
```

```python
b,g,r = cv.split(image) #splitting of image into each channels

cv.imshow('shankar', image)
cv.imshow('blue', b)
cv.imshow('green', g)
cv.imshow('red', r)

cv.waitKey(0)
cv.destroyAllWindows()
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
```

```python
image_merge = cv.merge((b,g,r)) # merge the three image up

cv.imshow('shankar_merged', image_merge)

cv.waitKey(0)
cv.destroyAllWindows()
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
```

# Add images

Merge two images(images have to be of same size). Also check weighted add, amount of weight for images to added.

```python
image1 = cv.imread('shankar_resized.jpeg', 1)
image2 = cv.imread('hand1.jpeg', 1)
```

```python
plt.imshow(image2)
```

```python
image1 = cv.resize(image1, (200, 200))
image2 = cv.resize(image2, (200,200))

image_add = cv.add(image1, image2) #images of same size can only be added

cv.imshow('imageadded', image_add)

cv.waitKey(0)
cv.destroyAllWindows()
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
```

```python
plt.imshow(image_add)
```

### Weighted add, priority the image

it take value: scource * alpha value the wieght for first image and scource2 * beta to the weight for second image, gamma is scalr value, destinatation and datatype.
