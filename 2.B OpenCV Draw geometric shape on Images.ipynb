```python
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
image = cv.imread("hand1.jpeg", 1)
```

```python
cv.imshow('Hand_image', image)

cv.waitKey(0)
cv.destroyAllWindows()
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
```

```python
plt.imshow(image)
```

```python
image.shape
```

```python
#draw a line on the image
#the inputs are (image,(start point (0,0), endpoint(186,272), colorofline in BGR format(blueline), thickness(1-10), )

image2 = cv.line(image, (0,0),(272,186), (0,0,255), 5)

#arrowed line

image3 = cv.arrowedLine(image, (0,0),(150,15), (0,0,255), 5)


cv.imshow('Hand_image', image3)

cv.waitKey(0)
cv.destroyAllWindows()
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
```

```python
#lets take different image

img = cv.imread("sunset1.jpeg", 1)
```

```python
plt.imshow(img)
```

```python
#draw rectange
#values:(image,topleft cordinate, bottomright, color(red),thickness(if we give -1 it will fill the rectange)

img2 = cv.rectangle(img,(110,25), (165,75), (0,0,255), 5) 
```

```python
plt.imshow(img2)
```

```python
cv.imshow('Sunset_image', img2)

cv.waitKey(0)
cv.destroyAllWindows()
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
```

```python
#draw circle
#values:(image, center of circle, radius, color, thickness)

img3 = cv.circle(img, (130, 50), 20,(0,0,255), 3)
```

```python
plt.imshow(img3)
```

```python
cv.imshow('Sunset_image', img3)

cv.waitKey(0)
cv.destroyAllWindows()
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
```

```python
#add test to image
#values:(image, test want to add, startpoint cord of text, fontface, fontsize, color of font, thinckness, linetype)

font = cv.FONT_HERSHEY_SIMPLEX
img4 = cv.putText(img,'sunset', (100,165), font, 1, (255,255,255), 1, cv.LINE_AA)
```

```python
cv.imshow('Sunset_image', img4)

cv.waitKey(0)
cv.destroyAllWindows()
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
cv.waitKey(1)
```
