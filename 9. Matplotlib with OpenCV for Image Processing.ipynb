# Using Matplotlib with OpenCV

```python
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
image = cv.imread('Lena_large.jpg', 1)
```

```python
plt.imshow(image)
plt.show()
```

```python
img = cv.cvtColor(image, cv.COLOR_BGR2RGB) #since openCV reads in BGR need to convert to RBG to see true color
```

```python
plt.imshow(img)
#plt.xticks([]), plt.yticks([]) # hide the x and y ticks showing the size.
plt.show()
```
